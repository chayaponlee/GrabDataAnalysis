```python
ratioRR = list()
ratioOT = list()
for x in range(14):
    ratio1 = (grab['Rides'].iloc[x])/(grab['Requests'].iloc[x])
    ratio2 = (grab['OnlineDrivers'].iloc[x])/(grab['TotalDrivers'].iloc[x])
    ratioRR.append(ratio1)
    ratioOT.append(ratio2)

grab['Rides/Requests'] = ratioRR
grab['Online/Total'] = ratioOT
```

```python
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    corr_mat = np.corrcoef(x,y)

    
    return corr_mat[0,1]
```

```python
def array_quant(x):
    avg = int(np.mean(x))
    med = int(np.median(x))
    var = np.mean((np.array(x)-avg)**2)
    std = np.sqrt(var)
    rmax = x.max()
    rmin = x.min()
    arr = np.array([avg,med,var,std,rmax,rmin])
    return(arr)

def array_quants(x):
    avg = np.mean(x)
    med = np.median(x)
    var = np.mean((np.array(x)-avg)**2)
    std = np.sqrt(var)
    rmax = x.max()
    rmin = x.min()
    arr = np.array([avg,med,var,std,rmax,rmin])
    return(arr)
    

```

```python
quantDuration = array_quant(grab['AvgDuration/Ride(min)'])
quantDistance = array_quant(grab['AvgDistance/Ride(km)'])
quantRequests = array_quant(grab['Requests'])
quantRides = array_quant(grab['Rides'])
quantNewDriver = array_quant(grab['NewDriver'])
quantOnlineDrivers = array_quant(grab['OnlineDrivers'])
quantRatioRR = array_quants(grab['Rides/Requests'])
quantRatioOT = array_quants(grab['Online/Total'])

```

```python
df_quant = pd.DataFrame(columns=['Metric','Requests','Rides','Fare','NewDriver','OnlineDrivers','AvgDistance/Ride(km)',
                                 'AvgDuration/Ride(min)','Rides/Requests','Online/Total'])
```

```python
df_quant['Metric'] = ['Mean','Median','Variance','StandardDeviation','Max14Days','Min14Days']
df_quant['Requests'] = quantRequests
df_quant['Rides'] = quantRides
df_quant['Fare'] = [15666,13737,27263997,5221,25843,8970]
df_quant['NewDriver'] = quantNewDriver
df_quant['OnlineDrivers'] = quantOnlineDrivers
df_quant['AvgDistance/Ride(km)'] = quantDistance
df_quant['AvgDuration/Ride(min)'] = quantDuration
df_quant['Rides/Requests'] = quantRatioRR
df_quant['Online/Total'] = quantRatioOT
```

```python
growth['Metric'] = ['Total in 1W','Total in 2W','+% Growth']
growth['Requests'] = [req,req1,(req1-req)/req * 100]
growth['Rides'] = [rides,rides1,(rides1-rides)/rides * 100]
growth['Fare'] = [fare,fare1,(fare1-fare)/fare * 100]
growth['NewDriver'] = [new,new1,(new1-new)/new * 100]
```
